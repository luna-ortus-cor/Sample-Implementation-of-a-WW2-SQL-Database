Business Rules:


World War 2 was fought between countries. Each country is identified by its name, and has several attributes, its army strength, number of armor, number of battleships, number of carriers, number of cruisers, number of destroyers, number of submarines, number of fighters, number of bombers, number of carriers lost, number of battleships lost, number of cruisers lost, number of destroyers lost, number of submarines lost, number of armor (tanks and SPGs) lost, number of aircraft lost, as well as the military and civilian casualties. Countries also have interactions with each other. A country can surrender to another country, can declare war on another country, and can have historical events with other countries (or themselves). For all of these we keep track of the date that it occurred, and for historical events we include a short description of the event. Events are identified by an ID, while for surrender we identify by the country surrendering. Declaration of war is identified by the two countries involved. 
 

Wars are also fought at places. A place is identified by its name, and it exists in a country. Each place exists only in one country, but a country can have multiple places in it.  

 
Each country also has commanders, who are identified by their name. They also have rank as an attribute. Each commander belongs to one country, but a country can have many commanders. 

 
Each country also has militaries, which are identified by the country they belong to, as well as the name of the military (e.g. German 6th Army, German 1st Panzer Division). They have type as an attribute (infantry, navy, air, armour, etc.). Each military belongs to one country, but a country can have many militaries.  


Militaries are commanded by commanders, and we take note of the start date, end date for which the military was under the command of a particular commander. A military may be commanded by multiple commanders at different dates, and a commander can similarly command multiple militaries at different dates.  

 
Each country also has ships, which are identified by their name (IJN Yukikaze, USS Enterprise, etc.). They also have type (destroyer, light cruisier, heavy cruiser, light aircraft carrier, aircraft carrier, battle cruiser, battleship, submarine, repair ship) as an attribute. A ship can be sunk by another ship, on a particular date. A ship may belong to a military. We take note of the start and end date for which a particular ship was in a particular military. A military can have multiple ships.  

 
Many operations in WW2 involved multiple armies on both sides (Axis and Allies). Thus, each Axis and Allied force is identified by a surrogate key, as well as a list of militaries in each force.  Each military can belong to more than one axis or allied force, and each axis of allied force can be made up of more than one military.  

 
A battle (synonymous with an operation), is identified by its id (a surrogate key), with name, description, and for each side (axis and allies), we keep track of number of carriers lost, number of battleships lost, number of cruisers lost, number of destroyers lost, number of submarines lost, number of armor (tanks and SPGs) lost, number of aircraft lost, and number of casualties (military) as attributes. Furthermore, a battle may be part of another larger battle, and likewise a large battle may comprise several smaller-scale battles.  

 
Finally, Axis, Allied forces may engage in battles at certain places. For these, we take note of the start and end date over which the battle occurred. The relationship for these is many to many to many to many.

```python
%%sql
DROP DATABASE IF EXISTS WW2;
CREATE DATABASE WW2;
USE WW2;

CREATE TABLE COUNTRY(
    NAME VARCHAR(30) PRIMARY KEY,
    MAXGSTR INTEGER(4),
    NUMARMOR INTEGER(4),
    NUMCV INTEGER(4),
    NUMBB INTEGER(4),
    NUMC INTEGER(4),
    NUMDD INTEGER(4),
    NUMSS INTEGER(4),
    NUMFIGHT INTEGER(4),
    NUMBOMB INTEGER(4),
    NUMARMORL INTEGER(4),
    NUMAIRCRAFTL INTEGER(4),
    NUMCVL INTEGER(4),
    NUMBBL INTEGER(4),
    NUMCL INTEGER(4),
    NUMDDL INTEGER(4),
    NUMSSL INTEGER(4),    
    MCASUALTY INTEGER(4),
    CCASUALTY INTEGER(4)
);

CREATE TABLE PLACE(
    NAME VARCHAR(100) PRIMARY KEY,
    CNAME VARCHAR(30),
    FOREIGN KEY (CNAME) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE SET NULL
);

CREATE TABLE COMMANDER(
    NAME VARCHAR(50) PRIMARY KEY,
    COMRANK VARCHAR(30) NOT NULL,
    CNAME VARCHAR(30),
    FOREIGN KEY (CNAME) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE SET NULL
);

CREATE TABLE SHIP(
    NAME VARCHAR(30) PRIMARY KEY,
    TYPE VARCHAR(20) NOT NULL,
    CNAME VARCHAR(30),
    SUNKBY VARCHAR(30),
    SUNKDATE DATE,
    FOREIGN KEY (CNAME) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE SET NULL,
    FOREIGN KEY (SUNKBY) REFERENCES SHIP(NAME) ON UPDATE CASCADE ON DELETE SET NULL
);

CREATE TABLE MILITARY(
    CNAME VARCHAR(30),
    NAME VARCHAR(50),
    TYPE VARCHAR(30) NOT NULL,
    PRIMARY KEY (CNAME, NAME),
    FOREIGN KEY (CNAME) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE CASCADE
);

CREATE TABLE HASS(
    SHIP VARCHAR(30),
    MCNAME VARCHAR(30),
    MNAME VARCHAR(50),
    SDATE DATE,
    EDATE DATE,
    PRIMARY KEY (SHIP, MCNAME, MNAME),
    FOREIGN KEY (SHIP) REFERENCES SHIP(NAME) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (MCNAME, MNAME) REFERENCES MILITARY(CNAME, NAME) ON UPDATE CASCADE ON DELETE CASCADE
);

CREATE TABLE COMMANDM(
    COMNAME VARCHAR(50),
    MCNAME VARCHAR(30),
    MNAME VARCHAR(50),
    SDATE DATE,
    EDATE DATE,
    PRIMARY KEY (COMNAME, MCNAME, MNAME),
    FOREIGN KEY (COMNAME) REFERENCES COMMANDER(NAME) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (MCNAME, MNAME) REFERENCES MILITARY(CNAME, NAME) ON UPDATE CASCADE ON DELETE CASCADE
);

CREATE TABLE SURRENDER(
    C1 VARCHAR(30),
    C2 VARCHAR(30),
    DDATE DATE,
    PRIMARY KEY (C1),
    FOREIGN KEY (C1) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (C2) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE CASCADE
);

CREATE TABLE WAR(
    C1 VARCHAR(30),
    C2 VARCHAR(30),
    DDATE DATE,
    PRIMARY KEY (C1, C2),
    FOREIGN KEY (C1) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (C2) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE CASCADE
);

CREATE TABLE EVENT(
    ID INTEGER(4),
    C1 VARCHAR(30),
    C2 VARCHAR(30),
    SDATE DATE,
    DESCR VARCHAR(1000) NOT NULL,
    PRIMARY KEY (ID),
    FOREIGN KEY (C1) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE SET NULL,
    FOREIGN KEY (C2) REFERENCES COUNTRY(NAME) ON UPDATE CASCADE ON DELETE SET NULL
);

CREATE TABLE ALLIES(
    ID INTEGER(4) PRIMARY KEY
);

CREATE TABLE AXIS(
    ID INTEGER(4) PRIMARY KEY
);

CREATE TABLE ALLIEDFORCE(
    MCNAME VARCHAR(30),
    MNAME VARCHAR(50),
    ALLIEDKEY INTEGER(4),
    PRIMARY KEY (MCNAME, MNAME, ALLIEDKEY),
    FOREIGN KEY (MCNAME, MNAME) REFERENCES MILITARY (CNAME, NAME) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (ALLIEDKEY) REFERENCES ALLIES (ID) ON UPDATE CASCADE ON DELETE CASCADE
);

CREATE TABLE AXISFORCE(
    MCNAME VARCHAR(30),
    MNAME VARCHAR(50),
    AXISKEY INTEGER(4),
    PRIMARY KEY (MCNAME, MNAME, AXISKEY),
    FOREIGN KEY (MCNAME, MNAME) REFERENCES MILITARY (CNAME, NAME) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (AXISKEY) REFERENCES AXIS (ID) ON UPDATE CASCADE ON DELETE CASCADE
);

CREATE TABLE BATTLE(
    ID INTEGER(4) PRIMARY KEY,
    NAME VARCHAR(50),
    DESCR VARCHAR(1000) NOT NULL,
    AXISNUMARMORL INTEGER(4),
    AXISNUMAIRCRAFTL INTEGER(4),
    AXISNUMCVL INTEGER(4),
    AXISNUMBBL INTEGER(4),
    AXISNUMCL INTEGER(4),
    AXISNUMDDL INTEGER(4),
    AXISNUMSSL INTEGER(4),  
    AXISCASUALTY INTEGER(4),
    ALLIEDNUMARMORL INTEGER(4),
    ALLIEDNUMAIRCRAFTL INTEGER(4),
    ALLIEDNUMCVL INTEGER(4),
    ALLIEDNUMBBL INTEGER(4),
    ALLIEDNUMCL INTEGER(4),
    ALLIEDNUMDDL INTEGER(4),
    ALLIEDNUMSSL INTEGER(4),  
    ALLIEDCASUALTY INTEGER(4),
    PARTOF INTEGER(4),
    FOREIGN KEY (PARTOF) REFERENCES BATTLE(ID) ON UPDATE CASCADE ON DELETE CASCADE
);

CREATE TABLE HAS(
    ID INTEGER(4),
    PNAME VARCHAR(100),
    AXISKEY INTEGER(4),
    ALLIEDKEY INTEGER(4),
    SDATE DATE,
    EDATE DATE,
    PRIMARY KEY (ID, PNAME, AXISKEY, ALLIEDKEY),
    FOREIGN KEY (ID) REFERENCES BATTLE(ID) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (PNAME) REFERENCES PLACE(NAME) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (AXISKEY) REFERENCES AXIS(ID) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (ALLIEDKEY) REFERENCES ALLIES(ID) ON UPDATE CASCADE ON DELETE CASCADE
);

CREATE TABLE ACCOUNTS(
    ID INTEGER(4) NOT NULL AUTO_INCREMENT,
    USERNAME VARCHAR(50) NOT NULL,
    PASSWORD VARCHAR(255) NOT NULL,
    EMAIL VARCHAR(255) NOT NULL,
    PRIMARY KEY (ID)
);

CREATE TABLE TABLESTATS(
    ID INTEGER(4) NOT NULL DEFAULT 0,
    NUMUSER INTEGER(4) DEFAULT 0,
    NUMCOUNTRY INTEGER(4) DEFAULT 0,
    NUMPLACE INTEGER(4) DEFAULT 0,
    NUMCOMMANDER INTEGER(4) DEFAULT 0,
    NUMMILITARY INTEGER(4) DEFAULT 0,
    NUMSHIP INTEGER(4) DEFAULT 0,
    NUMEVENT INTEGER(4) DEFAULT 0,
    NUMBATTLE INTEGER(4) DEFAULT 0,
    PRIMARY KEY (ID)
);

CREATE TRIGGER TRIGGER_NUMUSER1 AFTER INSERT ON ACCOUNTS FOR EACH ROW UPDATE TABLESTATS SET NUMUSER=(SELECT COUNT(*) FROM ACCOUNTS);
CREATE TRIGGER TRIGGER_NUMUSER2 AFTER DELETE ON ACCOUNTS FOR EACH ROW UPDATE TABLESTATS SET NUMUSER=(SELECT COUNT(*) FROM ACCOUNTS);
CREATE TRIGGER TRIGGER_NUMCOUNTRY1 AFTER INSERT ON COUNTRY FOR EACH ROW UPDATE TABLESTATS SET NUMCOUNTRY=(SELECT COUNT(*) FROM COUNTRY);
CREATE TRIGGER TRIGGER_NUMCOUNTRY2 AFTER DELETE ON COUNTRY FOR EACH ROW UPDATE TABLESTATS SET NUMCOUNTRY=(SELECT COUNT(*) FROM COUNTRY);
CREATE TRIGGER TRIGGER_NUMPLACE1 AFTER INSERT ON PLACE FOR EACH ROW UPDATE TABLESTATS SET NUMPLACE=(SELECT COUNT(*) FROM PLACE);
CREATE TRIGGER TRIGGER_NUMPLACE2 AFTER DELETE ON PLACE FOR EACH ROW UPDATE TABLESTATS SET NUMPLACE=(SELECT COUNT(*) FROM PLACE);
CREATE TRIGGER TRIGGER_NUMCOMMANDER1 AFTER INSERT ON COMMANDER FOR EACH ROW UPDATE TABLESTATS SET NUMCOMMANDER=(SELECT COUNT(*) FROM COMMANDER);
CREATE TRIGGER TRIGGER_NUMCOMMANDER2 AFTER DELETE ON COMMANDER FOR EACH ROW UPDATE TABLESTATS SET NUMCOMMANDER=(SELECT COUNT(*) FROM COMMANDER);
CREATE TRIGGER TRIGGER_NUMMILITARY1 AFTER INSERT ON MILITARY FOR EACH ROW UPDATE TABLESTATS SET NUMMILITARY=(SELECT COUNT(*) FROM MILITARY);
CREATE TRIGGER TRIGGER_NUMMILITARY2 AFTER DELETE ON MILITARY FOR EACH ROW UPDATE TABLESTATS SET NUMMILITARY=(SELECT COUNT(*) FROM MILITARY);
CREATE TRIGGER TRIGGER_NUMSHIP1 AFTER INSERT ON SHIP FOR EACH ROW UPDATE TABLESTATS SET NUMSHIP=(SELECT COUNT(*) FROM SHIP);
CREATE TRIGGER TRIGGER_NUMSHIP2 AFTER DELETE ON SHIP FOR EACH ROW UPDATE TABLESTATS SET NUMSHIP=(SELECT COUNT(*) FROM SHIP);
CREATE TRIGGER TRIGGER_NUMEVENT1 AFTER INSERT ON EVENT FOR EACH ROW UPDATE TABLESTATS SET NUMEVENT=(SELECT COUNT(*) FROM EVENT);
CREATE TRIGGER TRIGGER_NUMEVENT2 AFTER DELETE ON EVENT FOR EACH ROW UPDATE TABLESTATS SET NUMEVENT=(SELECT COUNT(*) FROM EVENT);
CREATE TRIGGER TRIGGER_NUMBATTLE1 AFTER INSERT ON BATTLE FOR EACH ROW UPDATE TABLESTATS SET NUMBATTLE=(SELECT COUNT(*) FROM BATTLE);
CREATE TRIGGER TRIGGER_NUMBATTLE2 AFTER DELETE ON BATTLE FOR EACH ROW UPDATE TABLESTATS SET NUMBATTLE=(SELECT COUNT(*) FROM BATTLE);
```

#### Query 1

Describe the query here in words (example of aggregate function)

Display the average number of forces deployed by the US and Japan in WW2. This can be used to show on average how much military strength the two had, and if there is significant disparity between the average value of one strength characteristic and the actual value, it may be worth investigating more why that is so. For example, the US had 16.1 million soldiers while Japan had 6.1 million, for an average value of 11.1 million that shows quite some disparity.

```python
%%sql
SELECT COUNT(*) AS NAME, AVG(MAXGSTR) AS MAXGSTR, AVG(NUMARMOR) AS NUMARMOR, AVG(NUMCV) AS NUMCV, AVG(NUMBB) AS NUMBB, AVG(NUMC) AS NUMC, AVG(NUMDD) AS NUMDD, AVG(NUMSS) AS NUMSS, AVG(NUMFIGHT) AS NUMFIGHT, AVG(NUMBOMB) AS NUMBOMB, AVG(NUMARMORL) AS NUMARMORL, AVG(NUMAIRCRAFTL) AS NUMAIRCRAFTL, AVG(NUMCVL) AS NUMCVL, AVG(NUMBBL) AS NUMBBL, AVG(NUMCL) AS NUMCL, AVG(NUMDDL) AS NUMDDL, AVG(NUMSSL) AS NUMSSL, AVG(MCASUALTY) AS MCASUALTY, AVG(CCASUALTY) AS CCASUALTY FROM COUNTRY WHERE NAME IN ('USA', 'Japan');
```

#### Query 2

Describe the query here in words (example of join)

Display all militaries existing on a certain date (here, existence is determined by whether it is currently being commanded by a commander). This is useful if we want to see all the existing militaries on a certain date, for example we want to find out what other militaries existed on the day of the attack on Pearl Harbour (7 Dec 1941) which we can then use to do other searches. 

```python
%%sql
SELECT * FROM MILITARY,COMMANDM WHERE COMMANDM.MCNAME=MILITARY.CNAME AND COMMANDM.MNAME=MILITARY.NAME AND (CAST('1941-12-07' AS DATE) BETWEEN COMMANDM.SDATE AND COMMANDM.EDATE);
```

#### Query 3

Describe the query here in words (example of nested query)

In WW2, ground strength was usually a good indicator of the industrial power and output of a country. Hence explaining why countries like the Soviet Union and USA dominated in later stages of the war. Hence, display all countries that Germany was stronger than (in terms of ground strength). 

```python
%%sql
SELECT * FROM COUNTRY WHERE COUNTRY.MAXGSTR<(SELECT MAXGSTR FROM COUNTRY WHERE COUNTRY.NAME LIKE '%GERMANY%');
```
